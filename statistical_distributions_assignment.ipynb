```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot
```

```python
mean, std = 2, 3
phi_X1 = np.random.normal(loc=mean, scale=std, size=100)

fig, ax = plt.subplots(2,2, figsize=(10, 5)) #number of rows=2, number of columns=2, frame size of the figure object

bins = 20

sns.histplot(phi_X1, kde=True, bins=bins, ax=ax[0,0])

s = pd.Series(phi_X1).value_counts(bins=bins).values.max()
ax[0,0].plot(mean*np.ones(s), range(s), color='r', linestyle='--')
ax[0,0].plot(mean*np.ones(s)+std, range(s), color='k', linestyle='--')
ax[0,0].plot(mean*np.ones(s)-std, range(s), color='k', linestyle='--')
ax[0,0].set_title('Normal Distribution', fontsize=14)
 
qqplot(phi_X1, line='45', ax=ax[0,1])

phi_X1 = (phi_X1 - mean)/std
sns.histplot(phi_X1, kde=True, bins=20, ax=ax[1,0])
s=pd.Series(phi_X1).value_counts(bins=bins).values.max()
ax[1,0].plot(0*np.ones(s), range(s), color='r', linestyle='--')
ax[1,0].plot(0*np.ones(s)+1, range(s), color='k', linestyle='--')
ax[1,0].plot(0*np.ones(s)-1, range(s), color='k', linestyle='--')
ax[1,0].set_title('Standard Normal Distribution', fontsize=14)

qqplot(phi_X1, line='45', ax=ax[1,1])

plt.tight_layout(w_pad=3)
plt.show()


#qqplot needs to be done after making the distribution "standard normal distribution"
```

2- In this question, you test whether the central limit theorem works. You generate 1000 variables with two normal distributions. You can determine the mean and standard deviation of these variables yourself. All you have to do is generate the first variable 50 times and averaged it each time. Generate the second variable 1000 times and averages this variable each time. Then plot the histogram of the averages of the two variables. Which of the variables has a mean distribution closer to the normal distribution? Do you think the Central Limit Theorem seems to have worked?

```python
mean=4.5
std=2
L1=[]
L2=[]

for i in range(50):
    X1=np.random.normal(loc=mean, scale=std, size=1000)
    L1.append(X1.mean())
    
for i in range(1000):
    X2=np.random.normal(loc=mean, scale=std, size=1000)
    L2.append(X2.mean())

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.histplot(L1, kde=True)

plt.subplot(1,2,2)
sns.histplot(L2, kde=True)

plt.show()

    
```

It important to understand the difference between the central limit theorem and the law of large numbers
