### Step 4: Running a Classifier
- Separate your dataset into feature set `X` and target variable `y`. Your target variable is `OnDiab`. 
- Split your dataset into train and test datasets, keep the test dataset size as 0.25 using `test_size` parameter of `train_test_split`. Set the random seed top 7 using `random_state` parameter of `train_test_split`. Make a stratified split.
- Train `KNearestNeighbor` classifer on your train dataset and print the score on the the test dataset. Set number of neighbors to 5.

```python
X = df.drop(['OnDiab'], axis=1)
y = df.OnDiab
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
```

##### stratify meaning: 
In the context of train-test splitting, the stratify parameter ensures that the class proportions in the target variable are preserved in both the training and testing datasets.

When you set stratify=y, where y represents the target variable, the train_test_split function will perform a stratified split. This means that it will allocate the data in a way that maintains the same class distribution as in the original dataset.

For example, let's say you have a binary classification problem where the target variable y has two classes: 0 and 1. If the original dataset has 80% of samples with class 0 and 20% with class 1, the stratified splitting will ensure that the training and testing datasets will also have a similar distribution of class 0 and class 1 samples. This is particularly useful when dealing with imbalanced datasets where one class has significantly fewer samples than the other.

By using stratify=y, you can have confidence that your training and testing datasets are representative of the original data distribution, helping to ensure that your model is trained and evaluated on a balanced and representative set of samples.

### Step 5 : Grid Search
- Import `GridSearchCV` from `sklearn.model_selection`
- Split your data into train and test datasets
- For `neighbors=1 to 30`, compute `GridSearchCV` for train dataset with kfold=10.
- Print the best cross validation score 
- Print the best parameter
- Print the test score

```python
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
mygridval={'n_neighbors': np.arange(1,30,2)}
```

```python
mygrid = GridSearchCV(KNeighborsClassifier(), param_grid=mygridval, cv=10, return_train_score=True)
```

```python
mygrid.fit(X_train, y_train)
```

```python
print("best cross validation score: {:.3f}".format(mygrid.best_score_))
print("best parameter: {}".format(mygrid.best_params_))
print("test score: {:.3f}".format(mygrid.score(X_test, y_test)))
```
