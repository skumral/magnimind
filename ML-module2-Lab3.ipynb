1) Load the breast_cancer dataset from skelarn (from sklearn.datasets import load_breast_cancer). Split the dataset into training and test datasets. Scale the dataset using minmaxscaler. Use KNeighborsClassifier classifier and report the score on the test dataset.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
score_train_scaled = knn.score(X_train_scaled, y_train)

score_test_scaled = knn.score(X_test_scaled, y_test)

print("score_train_scaled: ", score_train_scaled)
print("score_test_scaled: ", score_test_scaled)
```

2) Repeat Step 1 using pipelines and report the score.

```python
from sklearn.pipeline import make_pipeline
mypipeline=make_pipeline(MinMaxScaler(), KNeighborsClassifier())
```

```python
mypipeline.fit(X_train, y_train)
print('score_train_scaled:', mypipeline.score(X_train, y_train))
print('score_test_scaled:', mypipeline.score(X_test, y_test))
```

3) Use the pipeline object from Step 2 and make a grid search on parameter of number of neighbor. 

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

param_grid = {"kneighborsclassifier__n_neighbors": np.arange(1,30,2)}

mygrid = GridSearchCV(mypipeline, param_grid=param_grid, cv=10, return_train_score=True)

mygrid.fit(X_train, y_train)
```

```python
print("best mean cross-validation score: {:.3f}".format(mygrid.best_score_))
print("best parameters: {}".format(mygrid.best_params_))

print("test-set score: {:.3f}".format(mygrid.score(X_test, y_test)))
```

```python
import pandas as pd 
from sklearn.neighbors import KNeighborsRegressor
df = pd.read_csv('bike_day_raw.csv')

print(df.shape)
print(df.dtypes)
print(df.columns)
X=df.drop('cnt', axis=1)
y=df.cnt
#y=df.iloc[:,10]
```

```python
#Create a scatterplot of each feature against the target variable 
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(4, 3, figsize=(20,20))
for i, ax in enumerate(axes.ravel()):
    if i > 9:
        ax.set_visible(False)
        continue
    ax.plot(X.iloc[:, i], y, 'o', alpha=.5)
    ax.set_title("{}: {}".format(i, X.columns[i]))
    ax.set_ylabel("cnt")
```

```python
#Create alist of features that are numeric and not numeric    
numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
nonnumeric_features = ['season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
```

```python
#Create a pipeline of imputer and standard scaler for the numeric features
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
mypipe=make_pipeline(SimpleImputer(missing_values=np.nan, strategy='median'), StandardScaler())
```

```python
#Create a column transformer which uses the pipeline you created for numeric features 
#and a onehotencoder for the non-numeric features
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer #pipeline of pipelines
from sklearn.neighbors import KNeighborsRegressor

numeric_transformer = mypipe
nonnumeric_transformer = make_pipeline(OneHotEncoder(handle_unknown='ignore'))

preprocessor = ColumnTransformer(
    transformers = [
    ('num', numeric_transformer, numeric_features),
    ('nonnum', nonnumeric_transformer, nonnumeric_features)
])
```

```python
#Finally create a pipeline of column transformer and KNeighborsRegressor
final_pipe=Pipeline(steps = [
    ('column_transformer', preprocessor),
    ('knr', KNeighborsRegressor())
])
```

```python
#Split your data into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Report the score on the test dataset
final_pipe.fit(X_train, y_train)
```

```python
print("test score: ", final_pipe.score(X_test, y_test))
```
